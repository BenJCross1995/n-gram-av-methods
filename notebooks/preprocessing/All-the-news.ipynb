# All-the-news Fix

There was an issue where one of the authors in the metadata was not the same as in unknown and known

```python
import sys

from from_root import from_root

sys.path.insert(0, str(from_root("src")))

from utils import get_base_location
from read_and_write_docs import read_jsonl, read_rds, write_jsonl
```

```python
corpus      = "All-the-news"
data_type   = "training"

# Set NAS so can run on Windows laptop seamlessly
nas_base_loc = get_base_location()

# --- set your args (strings) ---
# nas_base_loc = "/Users/user/Documents/uni_work_offline"

known_loc = f"{nas_base_loc}/datasets/author_verification/{data_type}/{corpus}/known_raw.jsonl"
unknown_loc = f"{nas_base_loc}/datasets/author_verification/{data_type}/{corpus}/unknown_raw.jsonl"
metadata_loc = f"{nas_base_loc}/datasets/author_verification/{data_type}/metadata.rds"

save_loc = f"{nas_base_loc}/datasets/author_verification/{data_type}/{corpus}"
```

```python
metadata = read_rds(metadata_loc)
filtered_metadata = metadata[metadata['corpus'] == corpus]

known = read_jsonl(known_loc)
unknown = read_jsonl(unknown_loc)
```

## Check to see the mismatches

```python
bad_known_rows = filtered_metadata.loc[
    ~filtered_metadata["known_author"].isin(known["author"])
].copy()

bad_unknown_rows = filtered_metadata.loc[
    ~filtered_metadata["unknown_author"].isin(unknown["author"])
].copy()

print("Known author mismatches:")
print(bad_known_rows)

print("\nUnknown author mismatches:")
print(bad_unknown_rows)
```

## Fix the errors

```python
known.loc[
    (known["author"] == "Arriana McLymore"),
    "author"
] = "ARRIANA MCLYMORE"
```

```python
unknown.loc[
    (unknown["author"] == "Alberto Alerigi Jr."),
    "author"
] = "Alberto Alerigi Jr"

known.loc[
    (known["author"] == "Alberto Alerigi Jr."),
    "author"
] = "Alberto Alerigi Jr"
```

## Re-check the mismatches

```python
bad_known_rows = filtered_metadata.loc[
    ~filtered_metadata["known_author"].isin(known["author"])
].copy()

bad_unknown_rows = filtered_metadata.loc[
    ~filtered_metadata["unknown_author"].isin(unknown["author"])
].copy()

print("Known author mismatches:")
print(bad_known_rows)

print("\nUnknown author mismatches:")
print(bad_unknown_rows)
```

## Save

```python
write_jsonl(known, known_loc)
write_jsonl(unknown, unknown_loc)
```
