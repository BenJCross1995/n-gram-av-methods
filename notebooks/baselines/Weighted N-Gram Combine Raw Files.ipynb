# Combine the raw Weighted N-Gram Files

```python
import sys
import os

import pandas as pd

from from_root import from_root

sys.path.insert(0, str(from_root("src")))

from read_and_write_docs import read_excel_sheets, write_rds
from utils import list_xlsx_files
```

```python
models = ["gpt2"]

corpora = [
    "ACL", "Amazon", "Enron Cleaned", "Koppel's Blogs", "Perverted Justice",
    "Reddit", "StackExchange", "The Apricity", "The Telegraph", "TripAdvisor", "Wiki"
]

data_types = ["training", "test"]

weights = ["exp", "linear", "power"]

base_loc = "/Volumes/BCross/av_datasets_experiments/weighted_ngram_tracing"

output_file_name = "combined_metadata_raw.rds"
```

```python
total_combinations = len(data_types) * len(corpora) * len(models) * len(weights)

combination_num = 0

for d_type in data_types:
    for corpus in corpora:
        for model in models:
            for weight in weights:

                combination_num += 1

                folder_loc = f"{base_loc}/{d_type}/{corpus}/{model}/{weight}"
                save_loc = f"{base_loc}/{d_type}/{corpus}/{model}/{weight}_{output_file_name}"

                if os.path.exists(save_loc):
                    print(f"Skipping {combination_num} out of {total_combinations}: {d_type} | {corpus} | {model} | {weight} (save already exists)")
                    continue

                print(f"Working on {combination_num} out of {total_combinations}: {d_type} | {corpus} | {model} | {weight}")
                
                file_list = list_xlsx_files(folder_loc)
                
                result_list = []
                for file in file_list:
                    
                    metadata = read_excel_sheets(file, sheet_names=['metadata'])['metadata']
                    
                    result_list.append(metadata)
                    
                result_df = pd.concat(result_list)
                
                write_rds(result_df, save_loc)
```
