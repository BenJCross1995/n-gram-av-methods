```python
import os
import subprocess
from pathlib import Path
```

```python
# --- assume notebook is in: n-gram-av-methods/notebooks/script_testing/ ---
REPO_ROOT = Path.cwd().resolve().parents[2]

SCRIPT = REPO_ROOT / "scripts" / "score_ngrams_multi.py"

PYTHON = REPO_ROOT / "my_venv/bin/python"
```

```python

# --- set your args (strings) ---
KNOWN_LOC = "/Volumes/BCross/datasets/author_verification/test/Wiki/known_raw.jsonl"
UNKNOWN_LOC = "/Volumes/BCross/datasets/author_verification/test/Wiki/unknown_raw.jsonl"
METADATA_LOC = "/Volumes/BCross/datasets/author_verification/test/metadata.rds"
MODEL_LOC = "/Volumes/BCross/models/gpt2"
SAVE_LOC = "/Volumes/BCross/av_datasets_experiments/ngram_masking_logprobs_testing"
COMPLETED_LOC = "/Volumes/BCross/av_datasets_experiments/ngram_masking_logprobs_testing2"
CORPUS = "Wiki"
DATA_TYPE = "test"
KNOWN_DOC = "hodja_nasreddin_text_1"
UNKNOWN_DOC = "hodja_nasreddin_text_3"
```

```python
cmd = [
    str(PYTHON), "-u", str(SCRIPT),
    "--known_loc", KNOWN_LOC,
    "--unknown_loc", UNKNOWN_LOC,
    "--metadata_loc", METADATA_LOC,
    "--model_loc", MODEL_LOC,
    "--save_loc", SAVE_LOC,
    "--completed_loc", COMPLETED_LOC,
    "--corpus", CORPUS,
    "--data_type", DATA_TYPE,
    "--known_doc", KNOWN_DOC,
    "--unknown_doc", UNKNOWN_DOC,
]

print("Running:\n", " ".join(cmd))
```

```python
# run from repo root so relative paths inside the script behave as expected
proc = subprocess.run(
    cmd,
    cwd=str(REPO_ROOT),
    text=True,
    capture_output=True
)

print(proc.stdout)
if proc.returncode != 0:
    print(proc.stderr)
    raise RuntimeError(f"Script failed with exit code {proc.returncode}")
```
