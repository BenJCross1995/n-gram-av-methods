```python
import os
import subprocess
from pathlib import Path
```

```python
# --- assume notebook is in: n-gram-av-methods/notebooks/script_testing/ ---
REPO_ROOT = Path.cwd().resolve().parents[2]

SCRIPT = REPO_ROOT / "scripts" / "weighted_ngram_tracing_rough_linears.py"

PYTHON = REPO_ROOT / "my_venv/bin/python"
```

```python

# --- set your args (strings) ---
INPUT_DIR = "/Volumes/BCross/av_datasets_experiments/weighted_ngram_tracing_concat/test/Enron Cleaned/gpt2/linear_sample"
SAVE_DIR = "/Volumes/BCross/av_datasets_experiments/weighted_ngram_tracing_concat/test/Enron Cleaned/gpt2/linear_rough"
```

```python
cmd = [
    str(PYTHON), "-u", str(SCRIPT),
    "--input_dir", INPUT_DIR,
    "--save_dir", SAVE_DIR,
]

print("Running:\n", " ".join(cmd))
```

```python
# run from repo root so relative paths inside the script behave as expected
proc = subprocess.run(
    cmd,
    cwd=str(REPO_ROOT),
    text=True,
    capture_output=True
)

print(proc.stdout)
if proc.returncode != 0:
    print(proc.stderr)
    raise RuntimeError(f"Script failed with exit code {proc.returncode}")
```
