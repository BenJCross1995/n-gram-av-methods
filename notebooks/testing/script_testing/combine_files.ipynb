```python
import os
import sys
import subprocess
from pathlib import Path

from from_root import from_root

sys.path.insert(0, str(from_root("src")))

from read_and_write_docs import read_rds
```

```python
# --- assume notebook is in: n-gram-av-methods/notebooks/script_testing/ ---
REPO_ROOT = Path.cwd().resolve().parents[2]

SCRIPT = REPO_ROOT / "scripts" / "combine_files.py"

PYTHON = REPO_ROOT / "my_venv/bin/python"
```

```python
# --- set your args (strings) ---
MODE = "combine_xlsx"
COMBINE_LEVEL = "None"
DATA_LOC = "/Volumes/BCross/av_datasets_experiments/concat_ngram_masking_logprobs/test/Enron Cleaned/gpt2/raw_10"
SAVE_LOC = "/Volumes/BCross/av_datasets_experiments/testing/test.rds"
INPUT_FILES = "None"
SHEET_NAMES = "metadata"
```

```python
cmd = [
    str(PYTHON), "-u", str(SCRIPT),
    "--mode", MODE,
    "--combine_level", COMBINE_LEVEL,
    "--data_loc", DATA_LOC,
    "--save_loc", SAVE_LOC,
    "--input_files", INPUT_FILES,
    "--sheet_names", SHEET_NAMES,
    "--overwrite"
]

print("Running:\n", " ".join(cmd))
```

```python
# run from repo root so relative paths inside the script behave as expected
proc = subprocess.run(
    cmd,
    cwd=str(REPO_ROOT),
    text=True,
    capture_output=True
)

print(proc.stdout)
if proc.returncode != 0:
    print(proc.stderr)
    raise RuntimeError(f"Script failed with exit code {proc.returncode}")
```
