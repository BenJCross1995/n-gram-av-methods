```python
import os
import subprocess
from pathlib import Path
```

```python
# --- assume notebook is in: n-gram-av-methods/notebooks/script_testing/ ---
REPO_ROOT = Path.cwd().resolve().parents[2]

SCRIPT = REPO_ROOT / "scripts" / "weighted_ngram_tracing_concat.py"

PYTHON = REPO_ROOT / "my_venv/bin/python"
```

```python
# --- set your args (strings) ---
base_loc = "/Volumes/BCross"

if not os.path.exists(base_loc):
    print("Using Offline Location")
    base_loc = "/Users/user/Documents/uni_work_offline"
else:
    print("Using Volume Location")
    
# --- set your args (strings) ---
CORPUS = "Enron Cleaned"
DATA_TYPE = "test"
KNOWN_LOC = f"{base_loc}/datasets/author_verification/{DATA_TYPE}/{CORPUS}/known_raw.jsonl"
UNKNOWN_LOC = f"{base_loc}/datasets/author_verification/{DATA_TYPE}/{CORPUS}/unknown_raw.jsonl"
METADATA_LOC = f"{base_loc}/datasets/author_verification/{DATA_TYPE}/metadata.rds"
MODEL_LOC = f"{base_loc}/models/gpt2"
SAVE_LOC = f"{base_loc}/av_datasets_experiments/weighted_ngram_tracing_testing_concat"
COMPLETED_LOC = f"{base_loc}/av_datasets_experiments/weighted_ngram_tracing_testing_concat"
PROBLEM = "Kimberly.watson vs Larry.campbell"
WEIGHT = "exp"
BASE = "2.0"
```

```python
cmd = [
    str(PYTHON), "-u", str(SCRIPT),
    "--known_loc", KNOWN_LOC,
    "--unknown_loc", UNKNOWN_LOC,
    "--metadata_loc", METADATA_LOC,
    "--model_loc", MODEL_LOC,
    "--save_loc", SAVE_LOC,
    "--completed_loc", COMPLETED_LOC,
    "--corpus", CORPUS,
    "--data_type", DATA_TYPE,
    "--problem", PROBLEM,
    "--weight", WEIGHT,
    "--base", BASE
]

print("Running:\n", " ".join(cmd))
```

```python
# run from repo root so relative paths inside the script behave as expected
proc = subprocess.run(
    cmd,
    cwd=str(REPO_ROOT),
    text=True,
    capture_output=True
)

print(proc.stdout)
if proc.returncode != 0:
    print(proc.stderr)
    raise RuntimeError(f"Script failed with exit code {proc.returncode}")
```
